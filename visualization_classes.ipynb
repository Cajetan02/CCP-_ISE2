```python
# Bar Chart of Sales Data

class BarChart:
    @staticmethod
    def plot_sales(file_path):
        import pandas as pd
        import matplotlib.pyplot as plt

        data = pd.read_csv(file_path)
        plt.bar(data['Region'], data['Sales'], color='skyblue')
        plt.title('Sales by Region')
        plt.xlabel('Region')
        plt.ylabel('Sales')
        plt.show()

a=BarChart()
a.plot_sales('sales_data.csv')
```

```python
# Line Chart of Temperature Changes

class LineChart:
    @staticmethod
    def plot_temperature(file_path):
        import pandas as pd
        import matplotlib.pyplot as plt

        data = pd.read_csv(file_path)
        plt.plot(data['Day'], data['Temperature'], color='blue', marker='o')
        plt.title('Temperature Trend')
        plt.xlabel('Days')
        plt.ylabel('Temperature')
        plt.show()

a=LineChart()
a.plot_temperature('temperature_data.csv')
```

```python
# Scatter Plot of Student Scores

class ScatterPlot:
    @staticmethod
    def plot_scores(file_path):
        import pandas as pd
        import matplotlib.pyplot as plt

        data = pd.read_csv(file_path)
        plt.scatter(data['Hours Studied'], data['Scores Obtained'], color='red')
        plt.title('Study Hours vs Scores')
        plt.xlabel('Hours Studied')
        plt.ylabel('Scores Obtained')
        plt.show()

a=ScatterPlot()
a.plot_scores('student_scores.csv')
```

```python
class Histogram:
    @staticmethod
    def plot_scores_distribution(file_path):
        import pandas as pd
        import matplotlib.pyplot as plt
        data = pd.read_csv(file_path)
        plt.hist(data['Exam Scores'], bins=10, color='green', edgecolor='black')
        plt.title('Exam Scores Distribution')
        plt.xlabel('Scores')
        plt.ylabel('Frequency')
        plt.show()

a=Histogram()
a.plot_scores_distribution('exam_scores.csv')
```

```python
class Heatmap:
    @staticmethod
    def plot_correlation(file_path):
        import pandas as pd
        import matplotlib.pyplot as plt
        import seaborn as sns
        data = pd.read_csv(file_path)
        sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
        plt.title('Correlation Heatmap')
        plt.show()

a=Heatmap()
a.plot_correlation('correlation_data.csv')
```

```python
class WavePlot:
    @staticmethod
    def plot_sine_and_cosine():
        import pandas as pd
        import matplotlib.pyplot as plt
        import numpy as np
        x = np.linspace(0, 2 * np.pi, 100)
        plt.plot(x, np.sin(x), label='Sine')
        plt.plot(x, np.cos(x), label='Cosine')
        plt.title('Sine and Cosine Waves')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend()
        plt.show()

a=WavePlot()
a.plot_sine_and_cosine()
```

```python
class StackedBarChart:
    @staticmethod
    def plot_sales_by_category(file_path):
        import pandas as pd
        import matplotlib.pyplot as plt
        data = pd.read_csv(file_path)
        plt.bar(data['category'], data['region1'], label='Category A', color='blue') 
        plt.bar(data['category'],data['region2'], bottom=data['region1'], label='Category B', color='orange')
        plt.title('Stacked Bar Chart of Sales by Category')
        plt.xlabel('Regions')
        plt.ylabel('Sales')
        plt.legend(title='Category')
        plt.show()

a=StackedBarChart()
a.plot_sales_by_category('data.csv')
```

```python
class BoxPlot:
    @staticmethod
    def plot_price_distribution(file_path):
        import pandas as pd
        import matplotlib.pyplot as plt
        import seaborn as sns
        data = pd.read_csv(file_path)
        sns.boxplot(x='House Type', y='Price', data=data)
        plt.title('Price Distribution by House Type')
        plt.xlabel('House Type')
        plt.ylabel('Price')
        plt.show()
        
a=BoxPlot()
a.plot_price_distribution('house_prices.csv')
```

```python
class PairPlot:
    def __init__(self, file_path):
        import pandas as pd
        import matplotlib.pyplot as plt
        self.data = pd.read_csv(file_path)
    def plot_relationships(self):
        import pandas as pd
        import matplotlib.pyplot as plt
        import seaborn as sns
        sns.pairplot(self.data, hue='Species')
        plt.title('Pair Plot of Iris Dataset')
        plt.show()

a=PairPlot('iris.csv')
a.plot_relationships()
```

```python
class RegressionPlot:
    def __init__(self, dataset_path):
        import pandas as pd
        self.data = pd.read_csv(dataset_path)

    def plot_price_prediction(self):
        import matplotlib.pyplot as plt
        import seaborn as sns
        sns.regplot(x='SquareFootage', y='Price', data=self.data, color='blue',
                    line_kws={'color': 'red', 'linewidth': 2})
        plt.title('Regression Plot of Housing Prices')
        plt.xlabel('Square Footage')
        plt.ylabel('Price')
        plt.show()

a=RegressionPlot('regression.csv')
a.plot_price_prediction()
```

```python
class BubbleChart:
    def __init__(self, dataset_path):
        import pandas as pd
        self.data = pd.read_csv(dataset_path)

    def plot_profits(self):
        import matplotlib.pyplot as plt
        plt.scatter(self.data['Revenue'],self.data['Profit'], c='blue', alpha=0.5, edgecolors='black')
        plt.title('Bubble Chart of Company Profits')
        plt.xlabel('Revenue')
        plt.ylabel('Profit')
        plt.show()

a=BubbleChart('company_profits.csv')
a.plot_profits()
```

```python
class TimeSeriesPlot:
    def __init__(self, dataset_path):
        import pandas as pd
        self.data = pd.read_csv(dataset_path)

    def plot_energy_usage(self):
        import matplotlib.pyplot as plt
        plt.plot(self.data['Timestamp'], self.data['Energy Usage'], marker='o', linestyle='-', color='blue')
        plt.title('Time Series Plot of Energy Consumption')
        plt.xlabel('Timestamp')
        plt.ylabel('Energy Usage')
        plt.grid(True)
        plt.show()

a=TimeSeriesPlot('energy_consumption.csv')
a.plot_energy_usage()
```

```python
class ScatterMatrix:
    def __init__(self, dataset_path):
        import pandas as pd
        self.data = pd.read_csv(dataset_path)

    def plot_relationships(self):
        import matplotlib.pyplot as plt
        import seaborn as sns
        sns.pairplot(self.data, diag_kind='hist', palette='Set2')
        plt.show()

a=ScatterMatrix('car_specs.csv')
a.plot_relationships()
```
